## 1.3 Application Status Distribution

```python
# Status distribution from our analysis
status_data = {
    'Rejected': 3569,
    'Team Allocated': 3276,
    'Started': 767,
    'Dropped Out': 617,
    'Waitlisted': 109,
    'Applied': 105,
    'Withdraw': 86,
    'Rewards Award': 29
}

colors = ['#e74c3c','#27ae60','#27ae60','#e67e22','#95a5a6','#95a5a6','#95a5a6','#27ae60']

fig, ax = plt.subplots(figsize=(12, 6))
bars = ax.bar(status_data.keys(), status_data.values(), color=colors, edgecolor='white', linewidth=0.5)

for bar, val in zip(bars, status_data.values()):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 30,
            f'{val:,}', ha='center', va='bottom', fontsize=10, fontweight='bold')

ax.set_title('Application Status Distribution (n=8,558)', fontsize=14, fontweight='bold', pad=15)
ax.set_xlabel('Status', fontsize=11)
ax.set_ylabel('Count', fontsize=11)
ax.tick_params(axis='x', rotation=20)
plt.tight_layout()
plt.savefig('../outputs/week1_status_distribution.png', dpi=150, bbox_inches='tight')
plt.show()
print('\nPositive outcomes (Team Allocated + Started + Rewards Award): 4,072 (47.6%)')
print('Negative outcomes (Rejected + Dropped Out + Withdraw): 4,272 (49.9%)')
```

## 1.4 Opportunity Category Breakdown

```python
category_data = {
    'Internship': 5421,
    'Course': 2037,
    'Event': 545,
    'Competition': 425,
    'Engagement': 130
}

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Bar chart
cat_colors = ['#e74c3c', '#3498db', '#e67e22', '#27ae60', '#9b59b6']
bars = ax1.barh(list(category_data.keys()), list(category_data.values()),
                color=cat_colors, edgecolor='white')
for bar, val in zip(bars, category_data.values()):
    ax1.text(bar.get_width() + 50, bar.get_y() + bar.get_height()/2,
             f'{val:,} ({val/8558*100:.1f}%)', va='center', fontsize=10)
ax1.set_title('Records by Opportunity Category', fontweight='bold')
ax1.set_xlabel('Number of Records')
ax1.set_xlim(0, 6500)

# Pie chart
ax2.pie(category_data.values(), labels=category_data.keys(),
        colors=cat_colors, autopct='%1.1f%%', startangle=90)
ax2.set_title('Category Share of Total Records', fontweight='bold')

plt.suptitle('Opportunity Category Distribution', fontsize=13, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig('../outputs/week1_category_distribution.png', dpi=150, bbox_inches='tight')
plt.show()
```
