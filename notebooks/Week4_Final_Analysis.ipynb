## 4.4 Final Takeaways

### What We Learned

1. **Program design drives outcomes, not demographics.** Opportunity Category + Historical Rate account for 77% of model predictions. Student demographics (age, gender, country) contribute less than 10%.

2. **The internship funnel is the platform's core challenge.** 63% of applications go to internships; only 21.5% convert. This is a structural capacity problem, not a student quality problem.

3. **Speed is a signal.** Students who apply within 14 days of sign-up are twice as likely to be placed. The window of intent is short.

4. **One program sets the benchmark.** Career Essentials achieved 94.4% participation with 1,423 applicants — the highest volume and one of the highest success rates on the platform.

5. **The model works.** AUC-ROC of 0.9114 means the model correctly ranks participation likelihood 91% of the time. It can be deployed for real-time student nudging.
